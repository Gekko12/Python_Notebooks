### Out of all the available records how many days are within one standard deviation of average temperature


```python
mean_1 = df["Average temperature (°F)"].mean().round(2)
mean_1
```

```python
std_1 = df["Average temperature (°F)"].std().round(2)
std_1
```

```python
right_deviated = df["Average temperature (°F)"].round(2) <= (mean_1 + std_1)
left_deviated = df["Average temperature (°F)"].round(2) >= (mean_1 - std_1)
```

```python
df[right_deviated & left_deviated]
```

```python
q11 = len( df[right_deviated & left_deviated] )
q11
```
