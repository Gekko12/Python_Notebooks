## 500. Keyboard Row
Given an array of strings words, return the words that can be typed using letters of the alphabet on only one row of American keyboard like the image below.

In the American keyboard:
* the first row consists of the characters `"qwertyuiop"`,
* the second row consists of the characters `"asdfghjkl"`, and
* the third row consists of the characters `"zxcvbnm"`.

<img src="https://assets.leetcode.com/uploads/2018/10/12/keyboard.png" alt="keyboard" width=500 />

*Example 1:* <br>
Input: words = `["Hello","Alaska","Dad","Peace"]` <br>
Output: `["Alaska","Dad"]`<br>

*Example 2:* <br>
Input: words = `["omk"]` <br>
Output: `[]`

*Example 3:* <br>
Input: words = `["adsdf","sfd"]` <br>
Output: `["adsdf","sfd"]`
 

*Constraints:*
* 1 <= words.length <= 20
* 1 <= words[i].length <= 100
* words[i] consists of English letters (both lowercase and uppercase). 

```python
class Solution(object):
    def findWords(self, words):
        """
        :type words: List[str]
        :rtype: List[str]
        """
        key_rows = ["qwertyuiop", "asdfghjkl", "zxcvbnm"]
        res = []
        
        for word in words:  
            for row in key_rows:
                flag = True
                for ch in set(word):
                    if ch.lower() not in row:
                        flag = False
                        break
                if flag:
                    res.append(word)
                    break
                    
        return res
        

# main
obj = Solution()
words = ["Hello","Alaska","Dad","Peace"]
print(obj.findWords(words))
```

## 1470. Shuffle the Array
Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

Return the array in the form [x1,y1,x2,y2,...,xn,yn].


*Example 1:* <br>
Input: nums = [2,5,1,3,4,7], n = 3 <br>
Output: [2,3,5,4,1,7]  <br>
Explanation: Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

*Example 2:* <br>
Input: nums = [1,2,3,4,4,3,2,1], n = 4 <br>
Output: [1,4,2,3,3,2,4,1]

*Example 3:* <br>
Input: nums = [1,1,2,2], n = 2 <br>
Output: [1,2,1,2]
 

*Constraints:*
* $1 <= n <= 500$
* $nums.length == 2n$
* $1 <= nums[i] <= 10^3$

```python
class Solution(object):
    def shuffle(self, nums, n):
        """
        :type nums: List[int]
        :type n: int
        :rtype: List[int]
        """
        res = []
        for i in range(n):
            res.append(nums[i]);res.append(nums[i+n])
        return res
        
        
# main
obj = Solution()
nums = [2,5,1,3,4,7]
n = 3
print(obj.shuffle(nums, n))
```

## 350. Intersection of Two Arrays II
Given two integer arrays nums1 and nums2, return an array of their intersection. Each element in the result must appear as many times as it shows in both arrays and you may return the result in any order.

 
*Example 1:* <br>
Input: nums1 = [1,2,2,1], nums2 = [2,2] <br>
Output: [2,2]

*Example 2:* <br>
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4] <br>
Output: [4,9] <br>
Explanation: [9,4] is also accepted.
 

*Constraints:*
* 1 <= nums1.length, nums2.length <= 1000
* 0 <= nums1[i], nums2[i] <= 1000
 

*Follow up:*
* What if the given array is already sorted? How would you optimize your algorithm?
* What if nums1's size is small compared to nums2's size? Which algorithm is better?
* What if elements of nums2 are stored on disk, and the memory is limited such that you cannot load all elements into the memory at once?

```python
class Solution(object):
    def intersect(self, nums1, nums2):
        d1 = {key:0 for key in nums1}
        d2 = {key:0 for key in nums2}

        for key in nums1:
            d1[key] += 1
        for key in nums2:
            d2[key] += 1

        res = []

        for key, val in d1.items():
            try:
                itr = min(d1[key], d2[key])
                for i in range(itr):
                    res.append(key)
            except:
                pass
        return res
        
    def intersectSlow(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        lst = list(set(nums1) & set(nums2))
        len_ = len(lst)
        
        for idx in range(len_):
            mn = min(nums1.count(lst[idx]), nums2.count(lst[idx]))
            for cnt in range(mn-1):
                lst.append(lst[idx])
        
        return lst
                

#main
obj = Solution()
nums1 = [1,2,2,1,4,5]
nums2 = [2,2,2,2,1,4,1]
print(obj.intersect(nums1, nums2))
```

## 231. Power of Two
Given an integer n, return true if it is a power of two. Otherwise, return false.

An integer n is a power of two, if there exists an integer x such that n == 2x.

 
*Example 1:* <br>
Input: n = 1 <br>
Output: true <br>
Explanation: $2^0 = 1$ <br>

*Example 2:* <br>
Input: n = 16 <br>
Output: true <br>
Explanation: $2^4 = 16$

*Example 3:* <br>
Input: n = 3 <br>
Output: false

*Example 4:* <br>
Input: n = 4
Output: true

*Example 5:* <br>
Input: n = 5
Output: false
 

*Constraints:*
* $-2^{31} <= n <= 2^{31} - 1$
 

Follow up: Could you solve it without loops/recursion?

```python
from math import log2, ceil, floor

class Solution(object):
    def bitManipulation(self, n):
        if n <= 0:
            return False
        elif (n & ~(n-1)) == n:
            return True
        return False
    
        
    def isPowerOfTwo(self, n):
        """
        :type n: int
        :rtype: bool
        """
        return ceil(log2(n)) == floor(log2(n)) if n > 0 else False

    
    
# main
obj = Solution()
print(obj.isPowerOfTwo(16))
print(obj.bitManipulation(32))
```

## 1512. Number of Good Pairs
Given an array of integers nums, return the number of good pairs.

A pair (i, j) is called good if nums[i] == nums[j] and i < j.


**Example 1:** <br>
*Input:* nums = [1,2,3,1,1,3] <br>
*Output:* 4 <br>
*Explanation:* There are 4 good pairs (0,3), (0,4), (3,4), (2,5) 0-indexed.

**Example 2:** <br>
*Input:* nums = [1,1,1,1] <br>
*Output:* 6 <br>
*Explanation:* Each pair in the array are good.

**Example 3:** <br>
*Input:* nums = [1,2,3] <br>
*Output:* 0
 
**Constraints:**
* 1 <= nums.length <= 100
* 1 <= nums[i] <= 100

```python
from math import comb

class Solution:
    def numIdenticalPairs(self, nums):
        res = 0
        di = {num:0 for num in nums}
        
        for num in nums:
            di[num] += 1
        
        for k,v in di.items():
            if v > 1:
                res += comb(v, 2)    # combination of n by selcting 2 at a time ie. nC2 
        return res



#main
obj = Solution()
nums = [1,2,3,1,1,3]
print(obj.numIdenticalPairs(nums))
            
```

## 338. Counting Bits
Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

 
*Example 1:* <br>
Input: n = 2 <br>
Output: [0,1,1] <br>
Explanation: 
- 0 --> 0 
- 1 --> 1 
- 2 --> 10

*Example 2:* <br>
Input: n = 5 <br>
Output: [0,1,1,2,1,2] <br>
Explanation: 
- 0 --> 0 
- 1 --> 1 
- 2 --> 10
- 3 --> 11
- 4 --> 100
- 5 --> 101
 

*Constraints:*
* $0 <= n <= 10^5$
 

*Follow up:*
* It is very easy to come up with a solution with a runtime of O(n log n). Can you do it in linear time O(n) and possibly in a single pass?
* Can you do it without using any built-in function (i.e., like __builtin_popcount in C++)?

```python
from typing import List

class Solution:
    def countBits(self, n: int) -> List[int]:
        storage = [0] * (n+1)
        i = 1
        
        while i <= n:
            storage[i] = 1
            i *= 2
                
        i, j = 1, 1
        while i <= n:
            if not storage[i]:
                j += 1
                storage[i] = j
            else:
                j = 1
            i += 1
        
        return storage
    
    
# main
obj = Solution()
print([i for i in range(11)])
print(obj.countBits(10), "\nError")
```
