```python
import sqlite3 as sql
```

```python
help(sql)
```

```python
conn = sql.connect("mydb.db")
cur = conn.cursor()
```

```python
help(cur)
```

```python
drop_table = '''DROP TABLE IF EXISTS EMPLOYEE'''
cur.execute(drop_table)

create_table = '''CREATE TABLE EMPLOYEE (
                    NAME CHAR(60) NOT NULL,
                    EMPID INT,
                    ROLE CHAR(10),
                    PRIMARY KEY(EMPID))
                    '''
cur.execute(create_table)
```

```python
insert = '''INSERT INTO EMPLOYEE VALUES (?,?,?)'''
details = [("Ram", 101, "Developer"), ("Siya", 105, "IS-Admin"), ("Hanuman", 103, "Assistant Developer"), ("Alex", 110, "Assistant Developer"), ("Rita", 121, "Admin"), ("Gaurav", 109, "Assistant Developer")]

for tup in details:
    cur.execute(insert, tup)
```

```python
fetch  = cur.execute('''SELECT * FROM EMPLOYEE WHERE ROLE="Assistant Developer"''').fetchall()
fetch
```
