# <font color=hotpink> Matplotlib </font>

* Steps to be followed for creating effective visualizations.
    1. Don't bother much about the tool used for creating the visuals.
    2. Define what to communicate before we look at data.
    3. Identify the right chart or plot, which suits our story.
    4. Create the visual and verify if it is aligned with our story.
* There are many Python libraries used for Data Visualization. Few popular libraries are:
    * `matplotlib`: It is the most widely used Python Data Visualization library.
    * `seaborn`: It is used for generating informative statistical graphics. It is dependent on matplotlib.
    * `bokeh`: It is used for generating interactive plots, which can be accessed as JSON, HTML objects, or interactive web applications.
* In matplotlib, everything is organized in a hierarchy. At the top level, it consists of `matplotlib.pyplot` module. pyplot is used only for few activities such as figure creation.

````python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
````

## <font color=fe7401> Anatomy of Matplotlib figure </font>

* `Figure`: Whole area chosen for plotting, including axes, axis, and other artist elements.
* `Axes`: Area where data is plotted.
* `Axis`: Number-line like objects, which define graph limits.
* `Artist`: Every element on the figure is an artist.

<img src="./background/matplotlib_anatomy.gif"  alt="anatomy of matplotlib" width=400>

### <font color=blue> Figure </font>
* A figure is created using `figure` function of `pyplot` module, as shown below.
* Executing the below code doesn't display any figure
* `show` method can be used to view the created figure as shown below.
* The output simply shows the figure object.
* We will be able to view a picture only when a figure contains at least one `Axes` element.

````python
fig = plt.figure()
plt.show()
````

## <font color=blue> Axes </font>

* An Axes is the region of the figure, available for plotting data.
* An Axes object is associated with only **one** Figure.
* A Figure can contain one or more number of Axes elements.
* An Axes contains two Axis objects in case of 2D plots and three Axis objects in case of 3D plots.
* An Axes can be added to a figure using `add_subplot` methods.
    ``` 
    add_subplot(nrows, ncols, index) 
    
    When these argument values are less than 10, they all can be clubbed and passed as a single three-digit number.
    Hence, add_subplot(1, 1, 1) and add_subplot(111) are same.
    ```


````python
fig  = plt.figure()        # if we comment it out no figure will be shown
ax = fig.add_subplot(111)
plt.show()
````

### <font color=blue> Adjusting Figure Size </font>

* The default width and height of a figure are 6 and 4 inches respectively.
* We can change the size of a figure using `figsize` argument.
* For example, the expression `fig = plt.figure(figsize=(8,6))` generates a figure having 8 inches width and 6 inches height.

````python
fig = plt.figure(figsize = (2, 2))
ax = fig.add_subplot(111)
plt.show()
````

### <font color=blue> Setting Title and Axis Labels </font>

* `set` method can be used on created axes, to set various parameters such as xlabel, ylabel, title etc.

````python
fig = plt.figure(figsize = (4, 2))
ax = fig.add_subplot(111)
ax.set(title = 'My First Plot',
      xlabel = 'X-Axis', ylabel = 'Y-Axis',
      xlim = (0, 5), ylim = (0,10))
plt.show()
````

````python
# Setting an attribute can also be done with functions of the form set_<parameter_name>, as shown in below code.

fig = plt.figure(figsize = (2, 2))
ax = fig.add_subplot(111)

ax.set_title("XY Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xlim(left = 1, right = 4)
ax.set_ylim(bottom = 500, top = 0) 
# ax.set_ylim([100, 0])

plt.show()
````

### <font color=blue> Plotting Data </font>

* The `plot` is one of the functions used for plotting data points.
* plot function is called on the created axes object (this is good style).

* Plotting data or setting attributes can also be done by calling functions like `plot`, and `title` directly on `plt`.
* This would plot the data on axes, which is active *currently*. However, explicit is better than implicit. Hence prefer former style of plotting.


````python
fig = plt.figure(figsize = (2, 2))
ax = fig.add_subplot(111)
ax.set(title = "Polygon",
      xlabel = "X", ylabel = "Y",
      xlim = (0, 5), ylim = (0, 10),
      xticks = [1, 2, 5], xticklabels=["#", "##", "#####"])

x = [1, 2, 2, 1, 1]
y = [2, 4, 6, 8, 2]
plt.plot(x, y)
plt.show()
````

````python
# above one is better style

fig = plt.figure(figsize = (2, 2))

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y)

plt.title('My Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim(0,5)
plt.ylim(0,10)

plt.show()
````

### <font color=blue> Adding a Legend </font>

* `legend` function is called on axes object ax to produce a legend.
* The legend uses the `label`, provided to a line drawn using plot. 

````python
fig = plt.figure(figsize = (3, 3))
ax = fig.add_subplot(111)
ax.set(title='My Plot',
      xlabel='X-Axis', ylabel='Y-Axis',
      xlim=(0, 5), ylim=(0,10)) 

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, label='linear-growth')
plt.legend()
plt.show()
````

## <font color=fe7401> Line Plots </font>

* Line Plot is used to visualize a trend in data.
* Line Plot is also used to compare two variables.
* Line Plots are simple and effective in communicating.
* `plot` function is used for drawing Line plots.
    ```
    plot(x, y)
    # 'x' , 'y' : Data values representing two variables.
    ```

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = "Avg. Daily Temperature in 2018",
      xlabel = "Day", ylabel = "Temperature (in deg)",
      xlim = (0, 30), ylim = (25, 35))

days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
plt.plot(days, temp)
plt.show()
````

### <font color=blue> Common parameters of `plot` function </font>

* `color`: Sets the color of the line.
* `linestyle`: Sets the line style, e.g., solid, dashed, etc.
* `linewidth`: Sets the thickness of a line.
* `marker`: Chooses a marker for data points, e.g., circle, triangle, etc.
* `markersize`: Sets the size of the chosen marker.
* `label`: Names the line, which will come in legend.

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = "Avg. Daily Temperature in 2018",
      xlabel = "Day", ylabel = "Temperature (in deg)",
      xlim = (0, 30), ylim = (25, 35))

days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]

plt.plot(days, temp, color = 'green', linestyle = 'dashed', linewidth = 1, 
         label = "line", marker = "^", markerfacecolor = "red", markeredgecolor = "yellow")
plt.legend()  # also work with ax.legend() and ax.plot()
plt.show()
````

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = "Avg. Daily Temperature in 2018",
      xlabel = "Day", ylabel = "Temperature (in deg)",
      xlim = (0, 30), ylim = (25, 35))

days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]

ax.plot(days, temp, 'g^', days, temp, 'g-')
plt.show()
````

## <font color=fe7401> Scatter Plots </font>

* Scatter plot is very similar to Line Plot.
* Scatter Plot is used for showing how one variable is related with another.
* Scatter Plot consists of data points. If the spread of data points is linear, then two variables are highly correlated.
* `scatter` function is used for drawing scatter plots.
    ```
    scatter(x, y)
    # 'x', 'y' : Data values representing two variables.
    ```
 

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = "Avg. Daily Temperature in Jan 2018", 
      xlabel = "Day", ylabel = "Temperature (in deg)",
      xlim = (0, 30), ylim = (25, 35))

days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]

ax.scatter(days, temp)
plt.show()
````

### <font color=blue> Common Parameters of `scatter` </font>

* `c`: Sets color of markers.
* `s`: Sets size of markers.
* `marker`: Selects a marker. e.g: circle, triangle, etc
* `edgecolor`: Sets the color of lines on edges of markers.

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = "Avg. Daily Temperature in Jan 2018", 
      xlabel = "Day", ylabel = "Temperature (in deg)",
      xlim = (0, 30), ylim = (25, 35))

days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]

ax.scatter(days, temp, c=['#fe7401'], s=[80], edgecolors='yellow', marker = "*")
plt.show()
````

## <font color=fe7401> Bar Plot </font>

* Bar Plot is commonly used for comparing categories.
* It is also used to compare categories over a short period of time.
* `bar` and `barh` are used for plotting vertical and horizontal bar plots respectively.
    ```
    bar(x,height)
    # 'x' : x coordinates of bars.
    # 'height' : List of heights of each bar.
    
    barh(y, width)
    # 'y' : y coordinates of bars
    # 'width' : List of widths.
    ```


````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Quarterly Sales',
      xlabel = 'Quarter', ylabel = 'Sales (in millions)')

quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]

ax.bar(quarters, sales_2017)

ax.set_xticks(quarters)
ax.set_xticklabels(['Q1-2017', 'Q2-2017', 'Q3-2017'])

plt.show()
````

### <font color=blue> Common Parameters of `bar` </font>

* `color`: Sets the color of bars.
* `edgecolor`: Sets the color of the border line of bars.
* `width`: Sets the width of bars
* `align`: Aligns the bars w.r.t x-coordinates
* `label`: Sets label to a bar, appearing in legend.

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Quarterly Sales',
      xlabel = 'Quarter', ylabel = 'Sales (in millions)')

quarters = ['Q1-2017', 'Q2-2017', 'Q3-2017']
sales_2017 = [25782, 35783, 36133]

ax.bar(quarters, sales_2017, color = "red", edgecolor = "black", width = 0.5, label = "sales")
ax.legend()

plt.show()
````

### <font color=blue> Plotting Multiple Groups </font>

* Vertical bar plots are used for comparing more than one category at a time.

````python
fig = plt.figure(figsize = (5, 5))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Quarterly Sales',
      xlabel = 'Quarter', ylabel = 'Sales (in millions)')

quarters = [1, 2, 3]
x1_index = [0.8, 1.8, 2.8]      # this is required otherwise to align the bar, otherwise result would be different
x2_index = [1.2, 2.2, 3.2]
sales_2016 = [28831, 30762, 32178]
sales_2017 = [25782, 35783, 36133]

ax.bar(x1_index, sales_2016, color='yellow', 
       width=0.4, edgecolor='black', label='2016')   # width is in relation with (x2_index - x1_index) -> 0.4
ax.bar(x2_index, sales_2017, color='red',
       width=0.4, edgecolor='black', label='2017')
ax.set_xticks(quarters)
ax.set_xticklabels(["Q1", "Q2", "Q3"])
ax.legend()

plt.show()
````

### <font color=blue> Barplot Using `barh` </font>

* `barh` draws the bars horizontally.
* `height` parameter is used to adjust the height of each bar.
* Horizontal bar plots are used while comparing values of one category at a **time**

````python
fig = plt.figure(figsize = (3, 3))
ax = fig.add_subplot(111)
ax.set(title = 'Avg. Quarterly Sales',
      xlabel = 'Sales (in millions)', ylabel = 'Quarter')

quarters = [0.1, 0.5, 0.9]         # 0.1 + height(or)width + margin(user specified) -> 0.1 + 0.3 + 0.1
sales_2017 = [25782, 35783, 36133]

ax.barh(quarters, sales_2017, height = 0.3, color='red')    # heights means width of bar
ax.set_yticks(quarters)
ax.set_yticklabels(['Q1-2017', 'Q2-2017', 'Q3-2017'])

plt.show()
````

## <font color=fe7401> Pie Plot </font>

* Pie plot is effective in showing the proportion of categories.
* It is best suited for comparing fewer categories.
* In general, Pie Plot is used to highlight proportion of one or a group of categories.
    ```
    pie(x)
    # 'x' : sizes of portions, passed either as a fraction or a number.
    ```
* Common Parameters of pie
    * `colors`: Sets the colors of portions.
    * `labels`: Sets the labels of portions.
    * `startangle`: Sets the start angle at which portion drawing starts.
    * `autopct`: Sets the percentage display format of an area, covering portions.
    * `explode`: Used to fragment the pie chart with provided offsets.

````python
fig = plt.figure(figsize = (4, 4))
ax = fig.add_subplot(111)
ax.set(title='Avg. Quarterly Sales')

sales_2017 = [25782, 35783, 36133]

ax.pie(sales_2017, colors = [(1,0,0), "yellow", (0,1,0)],
       labels = ["Q1", "Q2", "Q3"], labeldistance = 1.2,
       shadow = True, startangle = 90,
       autopct = '%1.1f%%', pctdistance = 0.6,
       explode = [0, 0.15, 0.1])

plt.show()
````

## <font color=fe7401> Histogram </font>
* Histogram is used to visualize the spread of data of a distribution.
* `hist` function is used to plot a histogram.
    ```
    hist(x)
    # 'x' : Data values of a single variable.
    ```
* Common Parameters of `hist`
    * `color`: Sets the color of bars.
    * `bins`: Sets the number of bins to be used.
    * `normed`: Sets to True where bins display fraction and not the count.

````python
np.random.seed(42)
x = 60 + 10*np.random.randn(1000)

fig = plt.figure(figsize = (3, 3))
ax = fig.add_subplot(111)
ax.set(title = "Distribution of Student's Percentage",
      xlabel = "Percentage", ylabel = "Count")
ax.hist(x, bins = 20, color = "c")

plt.show()
````

## <font color=fe7401> Box Plot </font>

* Box plots are also used to visualize the spread of data.
* Box plots are used to compare distributions.
* Box plots can also be used to detect outliers.
    ```
    boxplot(x)
    # 'x' : list of values or list of list of values.
    ```
* Common Parameters of `boxplot`
    * `labels`: Sets the labels for box plots.
    * `notch`: Sets to True if notches need to be created around the median.
    * `bootstrap`: Number set to indicate that notches around the median are bootstrapped.
    * `vert`: Sets to False for plotting Box plots horizontally.

````python
np.random.seed(42)
x = 60 + 10*np.random.randn(1000)

fig = plt.figure(figsize = (5, 5))
ax = fig.add_subplot(111)
ax.set(title = "Box plot of Student's Percentage",
      xlabel = "Class", ylabel = "Percentage")
ax.boxplot(x, notch = True, labels = "A", bootstrap = 10000)

plt.show()

# multiple boxplots can be plot as shown below
# ax.boxplot([x, y, z], labels=['A', 'B', 'C'], notch=True, bootstrap=10000)
````

### <font color=blue> Using a Style </font>

* A specific style can be invoked with either of the two expressions shown below.
    ```
    plt.style.use('ggplot')
            or
    plt.style.context('ggplot'): Context manager for using style settings temporarily.
                                 Using this expression with a keyword, `with` is recommended
    ```
* Multiple style sheets can be used together in matplotlib.
* This provides the flexibility to compose two style sheets such as one for customizing colors and other for customizing element sizes.
    ```
    with plt.style.context(['dark_background', 'seaborn-poster']):
       ....
       ....
   ```

````python
with plt.style.context(['dark_background', 'seaborn-pastel']):
    fig = plt.figure(figsize = (5, 5))
    ax = fig.add_subplot(111)
    ax.set(title = 'Avg. Daily Temperature of Jan 2018',
      xlabel = 'Day', ylabel = 'Temperature (in deg)',
      xlim = (0, 30), ylim = (25, 35))
    
    days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
    temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
    
    ax.plot(days, temp, color = 'green', linestyle = '--', linewidth = 3)
    plt.show()
````

### <font color=blue>Creating a Custom Style</font>

* A style sheet is a text file having extension `.mplstyle`.
* All custom style sheets are placed in a folder, `stylelib`, present in the `config` directory of `matplotlib`.
* Use the below expression for knowing the Config folder.

## <font color=fe7401> Subplots </font>

* Now we will see how to create multiple plots in a single figure.
* `plt.subplot` is one of the functions used to create subplots.
    ```
    subplot(nrows, ncols, index)
    # 'index' is the position in a virtual grid with 'nrows' and 'ncols'
    # 'index' number varies from 1 to `nrows*ncols`.
    ```
* `subplot` creates the `Axes` object at `index` position and returns it.

````python
fig = plt.figure(figsize = (10, 8))

# The third argument, index value varied from 1 to 4, and respective subplots are drawn in row-major order.
axes1 = plt.subplot(2, 2, 2, title='Plot1')
axes2 = plt.subplot(2, 2, 4, title='Plot2')
axes3 = plt.subplot(2, 2, 3, title='Plot3')
axes4 = plt.subplot(2, 2, 1, title='Plot4')

plt.show()
````

````python
# Now let's try to create a figure with three subplots, where the first subplot spans all columns of first row.

fig = plt.figure(figsize = (8, 6))

ax1 = plt.subplot(2, 2, (1, 2), title = "Plot1")
ax1.set_xticks([])   # to remove all ticks of x and y
ax1.set_yticks([])

ax2 = plt.subplot(2, 2, 3, title = "Plot2")
ax3 = plt.subplot(2, 2, 4, title = "Plot3")
plt.show()
````

### <font color=blue> Subplots Using `GridSpec` </font>

* `GridSpec` class of `matplotlib.gridspec` can also be used to create Subplots.
* Initially, a grid with given number of rows and columns is set up.
* Later while creating a subplot, the number of rows and columns of grid, spanned by the subplot are provided as inputs to `subplot` function.

````python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize = (6, 4))
gd = gridspec.GridSpec(2, 2)

ax1 = plt.subplot(gd[0,:], title = "Plot1")
ax1.set_xticks([])  
ax1.set_yticks([])

ax2 = plt.subplot(gd[1, 0], title = "Plot2")
ax3 = plt.subplot(gd[1, 1], title = "Plot3")
plt.show()
````

````python
fig = plt.figure(figsize = (8, 6))
gd = gridspec.GridSpec(3, 3)

ax1 = plt.subplot(gd[0, :], title = "Plot 1")
ax1.set_xticks([])
ax1.set_yticks([])

ax2 = plt.subplot(gd[1, 0:2], title = "Plot 2")
ax2.set_xticks([])
ax2.set_yticks([])

ax3 = plt.subplot(gd[1:, 2], title = "Plot 3")
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = plt.subplot(gd[2, 0:2], title = "Plot 4")
ax4.set_xticks([])
ax4.set_yticks([])

plt.show()
````

## <font color=fe7401> Adding a Text </font>

* Text can be added to any part of the figure using `text` function.
    ```
    text(x, y, s)
    # 'x', 'y' represent x and y coordinates of a position.
    # 's' is the text or string to be written
    ```


````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.set(title='Writing Text',
      xlabel='X-Axis', ylabel='Y-Axis',
      xlim=(0, 5), ylim=(0, 9))

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

ax.scatter(x, y, c=['green'], s=[60], edgecolor='black')

for i in range(len(x)):
    str_temp = '({}, {})'.format(x[i], y[i])
    ax.text(x[i]-0.2, y[i]+0.2, str_temp, fontsize=16)
    
plt.show()
````

## <font color=fe7401> Matplotlib Backend </font>

* `matplotlib` can generate plots in different outputs.
* In general, backend refers to everything that occurs from the time of executing a plotting code to generating the figure.
* Backends are of two types.
    1. interactive backends: Graphical user interfaces like PyGTK, wxPython, Tkinter, qt4, etc.
    2. non-interactive backends: Image files such as PNG, JPEG, etc.

### <font color=blue> Choosing a Backend </font>

* The default backend chosen by matplotlib is available as backend setting in `matplotlibrc` file.
* If we want to alter the backend of many figures, change the value of backend setting.
* The below expression chooses WXAgg backend.
    ```
    backend : WXAgg
    ```
* We can also use use method if we want to go with a specific backend.
    ```
    import matplotlib
    matplotlib.use('WXAgg') 
    # Above expression must be used before importing pylot
    ```

## <font color=fe7401> Saving Figures </font>

* Once a backend is chosen, a matplotlib figure can be saved in any format supported by it.

````python
fig = plt.figure(figsize = (6, 4))
ax = fig.add_subplot(111)
ax.set(title = 'My First Plot',
      xlabel = 'X-Axis', ylabel = 'Y-Axis',
      xlim = (0, 5), ylim = (0,10))

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.savefig('myplot.png')
````

## <font color=fe7401> Common Pitfalls in Data Visualization </font>

1. Creating unlabelled plots.
2. Using 3-Dimensional charts. Don't prefer 3-D plots, unless they add any value over 2-D charts.
3. Portions of a pie plot do not sum up to meaningful number.
4. Showing too many portions in a single pie chart.
5. Bar charts not starting at zero.
6. Failing to normalize the data.
7. Adding extra labels, and fancy images.

## <font color=fe7401> Best Practices of Data Visualization </font>
1. Few of the best practices of Data Visualization are:
2. Display the data points on the plot, whenever required.
3. Whenever correlation is plotted, clarify that you have not established any cause of link between the variables.
4. Prefer labeling data objects directly inside the plot, rather than using legends.
5. Create a visualization, which stands by itself. Avoid adding extra text to tell more about visualization.
